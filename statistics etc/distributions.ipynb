```python
def binomial_pmf(n: int, p: float, x: int) -> float:    
    return (np.math.factorial(n) / (np.math.factorial(x) * np.math.factorial(n - x))) * (p ** x) * ((1 - p) ** (n - x))

n = 10
p = 0.5

x = np.arange(0, n+1)
y = [binomial_pmf(n, p, xi) for xi in x]

fig, ax = plt.subplots(facecolor='0.2')
ax.set_facecolor('black')

ax.bar(x, y, color='purple')

ax.set_xlabel("Number of successes")
ax.set_ylabel("Probability")
ax.set_title("Binomial Distribution (n = %d, p = %.2f)" % (n, p))

plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = list(map(lambda a: 1 / (1 + np.exp(-a)), x))

plt.figure(facecolor='grey',dpi = 100)
plt.axes().set_facecolor('grey')

plt.plot(x, y, color = 'black')
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")

plt.show()
```
