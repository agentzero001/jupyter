```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
C = np.sin(np.linspace(1,5,40))
NC = C + .5 * np.random.randn(40)
```

```python
plt.figure(facecolor='0.2')
plt.axes().set_facecolor('0.3')
plt.axis([-1,len(NC),min(NC)-0.1,max(NC)+0.1])
plt.plot(C, 'go', NC, 'r.')
plt.show()
```

```python
NC
```

```python
def movAvg(a,n):
    r = a.copy()
    for i in range(1, n):
        r += np.concatenate((r[:i] / i , a[:-i]))
        return r / n
```

```python
plt.figure(facecolor='0.2', figsize=(8,3))
plt.axes().set_facecolor('0.1')
plt.plot(NC, 'y', movAvg(NC,8), 'g*')
plt.show()
```
