```python
x = np.linspace(1,60,200)

fig = plt.figure(facecolor='grey')
ax = fig.gca(projection = '3d')
ax.set_facecolor('grey')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.scatter(np.sin(x),np.cos(x),x/2,c='black')
ax.scatter([4],[4],[4], c='red')
plt.show()
```

```python
x = np.linspace(-1.5,1.5,200)
x, y = np.meshgrid(x, x)
z = np.exp(-x*x-y*y)

fig = plt.figure(facecolor='grey')
ax = fig.gca(projection = '3d')
ax.set_facecolor('grey')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(x,y,z, cmap='Spectral')
```

```python
picture = np.zeros((200,200))
for x in range(200):
    for y in range(200):
        if x > 50 and y > 50 and y < 200 - x:
            picture[x,y] = 1
plt.figure(facecolor='0.2')
plt.imshow(picture,cmap='gray')
```

```python
filterX = np.zeros((3,3))
filterX[1,0] = -0.5
filterX[1,2] = 0.5

filterY = np.zeros((3,3))
filterY[0,1] = -0.5
filterY[2,1] = 0.5

filterX, filterY
```

```python
from scipy.signal import convolve2d
dx = convolve2d(picture, filterX)
dy = convolve2d(picture, filterY)
```

```python
def doubleplot(pic1, pic2, c='gray', s=(10,8)):
    fig,ax = plt.subplots(1,2, facecolor='0.2',figsize=s)
    ax[0].imshow(pic1, cmap=c)
    ax[1].imshow(pic2,cmap=c)
    plt.show()
```

```python
doubleplot(dx,dy)
```

```python
dx[45:60,45:55] #notice convolve2d will move the convolutionKernel in the opposite direction
```

```python
plt.figure(facecolor='0.2')
plt.imshow(np.sqrt(dx*dx+dy*dy), cmap='gray')
```
