```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def profit(x):
    return 0.5*x**2 - 4*x + 8

def derivative(x):
    return x - 4

x = np.linspace(-5, 10, 100)
y = profit(x)
y_prime = derivative(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4),facecolor= '0.3')


ax1.plot(x, y)
ax1.set_title("Profit function")
ax1.set_xlabel("x (number of cups)")
ax1.set_ylabel("P(x) (profit)")
ax1.axvline(x=4, color='black', linestyle='--',linewidth=1)
ax1.set_facecolor('grey')

ax2.plot(x, y_prime)
ax2.set_title("Derivative of profit function")
ax2.set_xlabel("x (number of cups)")
ax2.set_ylabel("P'(x) (rate of change of profit)")
ax2.axhline(y=0, color='black', linestyle='--',linewidth=1)
ax2.set_facecolor('grey')

plt.show()
```
