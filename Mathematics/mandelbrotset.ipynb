```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
def showMandelbrot():
    def Mandelbrot2(n):
        y, x = np.ogrid[-1.4:1.4:200j, -2.0:0.8:200j]
        z = x + y * 1j
        c = z
        for i in range(n):
            z = z * z + c
            z[z * np.conj(z) > 4] = 2.1
        return np.where(z * np.conj(z) > 4, 1, 0)  
    fig, axs = plt.subplots(3,3, facecolor='0.2', figsize=(10,10),sharey=True)    
    counter = 2
    for i in range(3):
        for j in range(3):
            axs[i][j].imshow(Mandelbrot2(counter), cmap='Spectral')
            axs[i][j].set_title('{}. Iteration'.format(counter))
            counter += 2 if counter < 9 else 4
```

```python
showMandelbrot()
```
