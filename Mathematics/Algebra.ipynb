```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
plt.rcParams['figure.facecolor'] = '.0'
plt.rcParams['axes.facecolor'] = '.1'
plt.rcParams['axes.labelcolor'] = 'white'
plt.rcParams['text.color'] = 'white'
plt.rcParams['xtick.color'] = 'green'
plt.rcParams['ytick.color'] = 'green'
plt.rcParams['axes.edgecolor'] = 'purple'
```

```python
c1 = lambda t: (np.sin(t), np.cos(t))
L  = np.linspace(0,np.pi*2,100)

circle = list(map(c1, L))
np_circle = np.array(circle)
plt.axes().set_facecolor('black')
plt.plot(np_circle[:,0], np_circle[:,1], color='orange')#, linestyle='dotted')
plt.axis('equal')

plt.show()
```

```python
rotation_matrix = np.array([[0, -1], [1, 0]])

vector1 = np.array((0,1))
pivot_point = np.array((2,1))
translated_vector1 = vector1 - pivot_point
rotated_translated_vector1 = translated_vector1 @ rotation_matrix
rotated_vector1 = rotated_translated_vector1 + pivot_point
```

```python
def label_vector(vector, color):
    plt.text(*(vector + .2), '({},{})'.format(vector[0],vector[1]), c=color)
```

```python
plt.axes().set_facecolor('black')
plt.scatter(*vector1, c='orange')
plt.scatter(*pivot_point, c='purple')
plt.scatter(*rotated_vector1, c='green')
plt.axhline(y=0,color='grey', zorder=0,alpha=.5)
plt.axvline(x=0, color='grey', zorder=0,alpha=.5)
label_vector(vector1, 'orange')
label_vector(pivot_point, 'purple')
label_vector(rotated_vector1, 'green')
plt.axis('equal')
plt.axis([-6,6,-6,6]);
```
