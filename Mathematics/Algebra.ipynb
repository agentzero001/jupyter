```python
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
plt.rcParams['figure.facecolor'] = '.0'
plt.rcParams['axes.facecolor'] = '.1'
plt.rcParams['axes.labelcolor'] = 'white'
plt.rcParams['text.color'] = 'white'
plt.rcParams['xtick.color'] = 'green'
plt.rcParams['ytick.color'] = 'green'
plt.rcParams['axes.edgecolor'] = 'purple'
plt.rcParams['text.usetex'] = False
```

```python
c1 = lambda t: (np.sin(t), np.cos(t))
L  = np.linspace(0,np.pi*2,100)

circle = list(map(c1, L))
np_circle = np.array(circle)
plt.axes().set_facecolor('black')
plt.plot(np_circle[:,0], np_circle[:,1], color='orange')#, linestyle='dotted')
plt.axis('equal')

plt.show()
```

```python
func = lambda x: 1/x
positive_values = np.linspace(.01,15,1000)
negative_values = np.linspace(-15,-.01,1000)

graph_pos = np.array(list(map(func, positive_values)))
graph_neg = np.array(list(map(func, negative_values)))
plt.axes().set_facecolor('black')
plt.axhline(y=0,color='grey', zorder=0,alpha=.5)
plt.axvline(x=0, color='grey', zorder=0,alpha=.5)
plt.plot(negative_values, graph_neg, color='orange')
plt.plot(positive_values, graph_pos, color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.title(r'f(x) = $\frac{1}{x}$')
plt.axis([-6,6,-6,6]);
plt.show()
```

```python
func = lambda x: (x**2 - 4) / (x**2 - 4*x)

positive_values2 = np.linspace(4.001, 10, 1000) 
positive_values = np.linspace(.01, 3.99, 1000)
negative_values = np.linspace(-15, -.01, 1000)

graph_pos = np.array(list(map(func, positive_values)))
graph_pos2 = np.array(list(map(func, positive_values2)))
graph_neg = np.array(list(map(func, negative_values)))
plt.axes().set_facecolor('black')
plt.axhline(y=0,color='grey', zorder=0,alpha=.5)
plt.axvline(x=0, color='grey', zorder=0,alpha=.5)
plt.plot(negative_values, graph_neg, color='orange')
plt.plot(positive_values, graph_pos, color='orange')
plt.plot(positive_values2, graph_pos2, color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.title(r'f(x) = $\frac{x^2 - 4}{x^2 - 4x}$',fontsize='16')
plt.axis([-6,6,-6,6]);
plt.show()
```

```python
func = lambda x: (9) / (x**2 - 9)

positive_values2 = np.linspace(3.001, 10, 1000) 
positive_values = np.linspace(-2.999, 2.999, 1000)
negative_values = np.linspace(-10, -3.001, 1000)

graph_pos = np.array(list(map(func, positive_values)))
graph_pos2 = np.array(list(map(func, positive_values2)))
graph_neg = np.array(list(map(func, negative_values)))
plt.axes().set_facecolor('black')
plt.axhline(y=0,color='grey', zorder=0,alpha=.5)
plt.axvline(x=0, color='grey', zorder=0,alpha=.5)
plt.plot(negative_values, graph_neg, color='orange')
plt.plot(positive_values, graph_pos, color='orange')
plt.plot(positive_values2, graph_pos2, color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.title(r'f(x) = $\frac{9}{x^2-9}$',fontsize='16')
plt.axis([-6,6,-6,6]);
plt.show()
```

```python
rotation_matrix = np.array([[0, -1], [1, 0]])

vector1 = np.array((0,1))
pivot_point = np.array((2,1))
translated_vector1 = vector1 - pivot_point
rotated_translated_vector1 = translated_vector1 @ rotation_matrix
rotated_vector1 = rotated_translated_vector1 + pivot_point
```

```python
def label_vector(vector, color):
    plt.text(*(vector + .2), '({},{})'.format(vector[0],vector[1]), c=color)
```

```python
plt.axes().set_facecolor('black')
plt.scatter(*vector1, c='orange')
plt.scatter(*pivot_point, c='purple')
plt.scatter(*rotated_vector1, c='green')
plt.axhline(y=0,color='grey', zorder=0,alpha=.5)
plt.axvline(x=0, color='grey', zorder=0,alpha=.5)
label_vector(vector1, 'orange')
label_vector(pivot_point, 'purple')
label_vector(rotated_vector1, 'green')
plt.axis('equal')
plt.axis([-6,6,-6,6]);
```
