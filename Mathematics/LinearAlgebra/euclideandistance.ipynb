```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
plt.rcParams['figure.facecolor'] = '0.3'
```

```python
%matplotlib inline
```

```python
A = np.array([2, 7])
B = np.array([8, 6])

distance = round(np.linalg.norm(B - A), 4)
distance2 = np.sqrt((B[0] - A[0])**2 + (B[1] - A[1])**2)
```

```python
B - A
```

```python
distance2, distance
```

```python
plt.figure()
plt.gca().set_facecolor('0.1')

plt.plot([A[0], B[0]], [A[1], B[1]], c='green', zorder=2, label='distance = {}'.format(distance))
plt.scatter(A[0], A[1], c='r', marker='o', zorder=3, label='A = ({}, {})'.format(A[0], A[1]))
plt.scatter(B[0], B[1], c='yellow', marker='o', zorder=3, label='B = ({}, {})'.format(B[0], B[1]))

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.legend(loc='upper right')
plt.legend().get_frame().set_alpha(0.3)
plt.grid(alpha=0.3,zorder=1)
```

```python
A = np.array([5, 5])
t = np.linspace(0, 4*np.pi-4*np.pi/100, 100)
points = np.zeros((100, 2))
points[:, 0] = A[0] + 5 * np.cos(t)
points[:, 1] = A[1] + 5 * np.sin(t)
points += np.random.randn(100,2)*.5
```

```python
distances = np.linalg.norm(points - A, axis=1)
distances
```

```python
plt.figure()
plt.gca().set_facecolor('0.0')

plt.scatter(A[0], A[1], c='y', marker='o', zorder=3,s=90)
plt.scatter(points[:, 0], points[:, 1], c='red', marker='o', zorder=3,s=10)
for i in range(len(points)):
    plt.plot([A[0], points[i, 0]], [A[1], points[i, 1]],c='green',linewidth=.2, zorder=2)
    
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-2, 12)
plt.ylim(-2, 12)
plt.grid(alpha=0.3,zorder=1)
plt.show()
```

```python
A2 = np.array([1, 2, 3])
B2 = np.array([4, 5, 6])

xs = [A2[0], B2[0]]
ys = [A2[1], B2[1]]
zs = [A2[2], B2[2]]

dist = np.sqrt(np.sum(np.square(np.subtract(A2, B2))))
dist2 = np.sqrt((B2[0] - A2[0])**2 + (B2[1] - A2[1])**2 + (B2[2] - A2[2])**2)
```

```python
dist, dist2
```

```python
fig = plt.figure()
fig.patch.set_facecolor('0.2')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('0.2')

ax.scatter(A2[0], A2[1], A2[2], c='b', marker='o')
ax.scatter(B2[0], B2[1], B2[2], c='r', marker='o')
ax.plot(xs, ys, zs, c='black')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(0, 7)
ax.set_ylim(0, 7)
ax.set_zlim(0, 7)

plt.show()
```

```python
A1 = np.array([5,5,5])

radius = 6
n = 100
theta = np.random.uniform(0, np.pi, n)
phi = np.random.uniform(0, 2*np.pi, n)

x = A1[0] + radius * np.sin(theta) * np.cos(phi)
y = A1[1] + radius * np.sin(theta) * np.sin(phi)
z = A1[2] + radius * np.cos(theta)

points_ = np.array([x, y, z])
points_.shape
```

```python
fig = plt.figure()
fig.patch.set_facecolor('0.2')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('0.2')



ax.scatter(A1[0], A1[1], A1[2], c='black', marker='o',s=300)
ax.scatter(points_[0], points_[1], points_[2], c='r', marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)

plt.show()
```
