```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.facecolor'] = '.1'
plt.rcParams['axes.facecolor'] = 'black'
```

```python
#verification of the solution for the system of linear equations
# 5*x + y = 3
# 2*x - y = 4
```

```python
coefficients = np.array([[5, 1], [2, -1]])
constants = np.array([3, 4])
solution = np.linalg.solve(coefficients, constants)
solution
```

```python
equation1 = list(map(lambda x: -5 * x + 3, np.arange(-20,20,1)))
equation2 = list(map(lambda x: 2 * x - 4, np.arange(-20,20,1)))
```

```python
plt.plot(np.arange(-20,20,1), equation1, color='green')
plt.plot(np.arange(-20,20,1), equation2, color='green')
plt.axis([-7,7,-7,7])
plt.axhline(-2, alpha=.3, color='red',linestyle='dashed')
plt.axvline(1,alpha=.3, color='red', linestyle='dashed')
plt.axhline(0, alpha=.1, color='white')
plt.axvline(0,alpha=.1, color='white')
#plt.grid(alpha=.1)
```

```python
X = np.array([10, 14])
Y = np.array([10, 12])
Z = np.array([[10], [15]])
```

```python
fig = plt.figure(facecolor='.1')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('.1')
ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
ax.set_zlabel('x_3')
ax.set_xlim(5, 15)
ax.set_ylim(5, 15)
ax.set_zlim(10, 20)

ax.plot_surface(X, Y, Z, alpha=0.5, color='black')
plt.show()
```
