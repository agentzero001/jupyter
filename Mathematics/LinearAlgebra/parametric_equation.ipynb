```python
import numpy as np
import matplotlib.pyplot as plt 

customizations = {
    'figure.facecolor': '.0',
    'axes.facecolor': '.1',
    'axes.labelcolor': 'white',
    'text.color': 'white',
    'xtick.color': 'green',
    'ytick.color': 'green',
    'axes.edgecolor': 'grey',
    'text.usetex': False
}
plt.rcParams.update(customizations)
```

```python
def label_vector(vector, color):
    ax.text(*(vector + .2), '({},{})'.format(vector[0], vector[1]), c=color)
```

$\vec{p} = \begin{bmatrix} 5 \\ 3 \end{bmatrix} \quad \vec{q} = \begin{bmatrix} 2 \\ 6 \end{bmatrix} \quad$
$\mathbf{\alpha}(t) = \begin{bmatrix} 3t+2 \\ -3t+6 \end{bmatrix} \quad$
$\mathbf{\alpha}'(t) = \begin{bmatrix} 3 \\ -3 \end{bmatrix} \quad$
$\lVert \mathbf{\alpha}'(t) \rVert = 3\sqrt{2}$

```python
t = np.linspace(0,1,1000)
get_values = lambda f, t: list(map(f, t))

p, q = p_q = np.array([[5,3],
                       [2,6]])

#x_coords = get_values(lambda t: (1-t) * p[0] + t * q[0], t)
#y_coords = get_values(lambda t: (1-t) * p[1] + t * q[1], t)
x_coords = get_values(lambda t:  3 * t + 2, t)
y_coords = get_values(lambda t: -3 * t + 6, t)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ax.axis([0,10,0,10])
ax.scatter(p_q[:,0], p_q[:,1], c=['orange', 'purple'])

for vector, color in zip(p_q,['orange', 'purple']):
    label_vector(vector, color)
    
ax.plot(x_coords,y_coords, zorder=0);
```

```python
c1 = lambda t: (np.sin(t*2), np.cos(t*3))
L  = np.linspace(0, 2*np.pi - 2*np.pi / 1000,1000)

circle = list(map(c1, L))
np_circle = np.array(circle)
plt.axes().set_facecolor('black')
plt.plot(np_circle[:,0], np_circle[:,1], color='grey')
plt.axis('equal')
plt.show()
```
