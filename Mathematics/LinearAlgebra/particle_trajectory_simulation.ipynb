```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.facecolor'] = '.2'
plt.rcParams['axes.facecolor'] = '.2'
```

```python
num_points = 100
np.random.seed(20)
noise = np.random.normal(loc = 3,scale=.3, size=(num_points)) * 2
noise2 = (np.sin(np.random.uniform(0, 2*np.pi, num_points)) + 1) * .2
```

```python
x = np.zeros(num_points) + 2 * noise2
y = np.linspace(-20, 20, num_points)
z = np.zeros(num_points) + 6 + noise


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, c='black', alpha=.6)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(-1,10)
ax.set_ylim(-1,10)
ax.set_zlim(-1,20)




plt.show()
```
