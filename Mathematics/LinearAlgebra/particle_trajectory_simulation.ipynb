```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.facecolor'] = '.2'
plt.rcParams['axes.facecolor'] = '.2'
```

```python
num_points = 50
np.random.seed(20)
noise = np.random.normal(loc = 3,scale=.3, size=(num_points))
noise2 = np.random.normal(loc = 8,scale=1, size=(num_points))
```

```python
start_point = np.array([0, 0, 8]) 
end_point = np.array([20, 20, 8])  

x = np.linspace(start_point[0], end_point[0], num_points)
y = np.linspace(start_point[1], end_point[1], num_points)
z = np.linspace(start_point[2], end_point[2], num_points)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x*np.random.normal(.5), noise, z, c='black', alpha=.6)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(-1,10)
ax.set_ylim(-1,10)
ax.set_zlim(-1,10)

plt.show()
```
