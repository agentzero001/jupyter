```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 3 * np.random.rand(20,1)
y = 2 * x + 0.4*np.random.randn(20,1) +1

mean_x = np.mean(x)
mean_y = np.mean(y)


numer = 0
denom = 0
for i in range(len(x)):
    numer += (x[i] - mean_x) * (y[i] - mean_y)
    denom += (x[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)

prediction = lambda x: b0 + b1 * x

print('prediction at x = 3: {}'.format(prediction(3)))

x_plot = np.linspace(0, 10, 100)
y_plot = b0 + b1*x_plot

plt.figure(facecolor='0.3')
plt.axes().set_facecolor('.2')
plt.axis([0,4,0,10])
plt.axvline(x=3, color='black', linestyle='--',linewidth=1)
plt.axhline(y=prediction(3), color='black', linestyle='--',linewidth=1)

plt.scatter(x, y, c='black')
plt.plot(x_plot,y_plot, c='red')
plt.show()
```

```python
class LinearRegressionModel:
    '''The Input Data has to be a list of tuples representing coordinate pairs'''
    
    get_coords = lambda self, D: ([point[0] for point in D], [point[1] for point in D])
    predict = lambda self, x: self.b0 + self.b1 * x
        
    def __init__(self, data):
        self.data = data
        self.x, self.y = self.get_coords(data)
        self.b0, self.b1 = self.get_coeffs(self.x,self.y)
        print('Training Succeeded: b0 = {}, b1 = {}'.format(self.b0,self.b1)) 
        
    @staticmethod   
    def get_coeffs(x, y):
        mean_x = np.mean(x)
        mean_y = np.mean(y)
        numer = 0
        denom = 0
        for i in range(len(x)):
            numer += (x[i] - mean_x) * (y[i] - mean_y)
            denom += (x[i] - mean_x) ** 2
        b1 = numer / denom
        b0 = mean_y - (b1 * mean_x)
        return b0, b1
        
    def plot(self, type_, x=None):
        plt.figure(facecolor='0.4')
        plt.axes().set_facecolor('black')
        plt.axis([min(self.x)-10,max(self.x)+10,min(self.y)-10,max(self.y)+10]) 
        plt.axvline(x=0, color='grey', linewidth=1)
        plt.axhline(y=0, color='grey', linewidth=1)        
        plt.grid(zorder = 1, alpha=.3)
        x_plot = np.linspace(-100, 100, 100)
        y_plot = self.b0 + self.b1*x_plot
        if type_ == 'data':
            plt.scatter(self.x, self.y, c='green',zorder=3)
            plt.show()
        if type_ == 'graph':
            plt.plot(x_plot,y_plot, c='red',zorder=2)
            plt.show()
        if type_ == 'graph and data':
            plt.scatter(self.x,self.y, c='green',zorder=3)
            plt.plot(x_plot,y_plot, c='red',zorder=2)
            plt.show()
        if type_ == 'predict':
            print('prediction at x = {} : {} '.format(x,self.predict(x)))       
            plt.scatter(x, self.predict(x), zorder = 4,s = 100, marker='o', color = 'grey')
            plt.axvline(x=x, color='grey', linestyle='--',linewidth=1)
            plt.axhline(y=self.predict(x), color='grey', linestyle='--',linewidth=1)
            plt.scatter(self.x,self.y, c='green',zorder=3)
            plt.plot(x_plot,y_plot, c='red',zorder=2)
            plt.show()
```

```python
my_data = [(1,1),(4,2),(2,4),(5,6),(6,5),(3,4),(10,10),(31,70)]

myModel = LinearRegressionModel(my_data)
```
