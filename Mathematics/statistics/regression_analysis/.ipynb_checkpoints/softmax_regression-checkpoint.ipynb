```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['figure.facecolor'] = '.0'
plt.rcParams['axes.facecolor'] = '0'
plt.rcParams['axes.labelcolor'] = 'grey'
plt.rcParams['text.color'] = 'grey'
plt.rcParams['xtick.color'] = 'green'
plt.rcParams['ytick.color'] = 'green'
plt.rcParams['axes.edgecolor'] = 'purple'
```

$\sigma(x) = \frac{1}{1 + e^{-x}}$

```python
sigma = lambda x: 1 / (1 + np.e**(-x))
x_ = np.linspace(-10, 10, 1000)
```

```python
plt.plot(x_, sigma(x_), color='grey')
plt.axhline(0, color='grey', linewidth=0.5)
plt.axvline(0, color='grey', linewidth=0.5)
plt.grid(alpha=.2)
```

```python
y = lambda x: 1 if sigma(x) > .5 else 0
```
