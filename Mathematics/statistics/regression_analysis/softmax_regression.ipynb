```python
sigma = lambda x: 1 / (1 + np.e**(-x))

def loss2(X, y, w):
    sum_ = 0
    for x, y_i in zip(X, y):
        sum_ += -y_i * np.log(sigma(x @ w)) - (1 - y_i) * np.log(1-sigma(x @ w))
    return sum_ / X.shape[0]
```

```python
num_obs = 20
num_features = 30

w = np.ones(num_features)
x = np.linspace(-3, 3, num_features)
y = np.r_[np.zeros(num_obs//2), np.ones(num_obs//2)]
```

```python
X = np.array([x] * num_obs)
X.shape
```

```python
y #the most trivial case
```

```python
loss2(X, y, w)
```
