```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
plt.rcParams['figure.facecolor'] = '0.3'
```

```python
%matplotlib inline
```

```python
A = np.array([2, 7])
B = np.array([8, 6])

distance = round(np.linalg.norm(B - A), 4)
distance2 = np.sqrt((B[0] - A[0])**2 + (B[1] - A[1])**2)
```

```python
distance2, distance
```

```python
plt.figure()
plt.gca().set_facecolor('0.1')

plt.plot([A[0], B[0]], [A[1], B[1]], c='green', zorder=2, label='distance = {}'.format(distance))
plt.scatter(A[0], A[1], c='r', marker='o', zorder=3, label='A = ({}, {})'.format(A[0], A[1]))
plt.scatter(B[0], B[1], c='yellow', marker='o', zorder=3, label='B = ({}, {})'.format(B[0], B[1]))

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.legend(loc='upper right')
plt.legend().get_frame().set_alpha(0.3)
plt.grid(alpha=0.3,zorder=1)
```

```python
A2 = np.array([1, 2, 3])
B2 = np.array([4, 5, 6])

xs = [A2[0], B2[0]]
ys = [A2[1], B2[1]]
zs = [A2[2], B2[2]]

dist = np.sqrt(np.sum(np.square(np.subtract(A2, B2))))
dist2 = np.sqrt((B2[0] - A2[0])**2 + (B2[1] - A2[1])**2 + (B2[2] - A2[2])**2)
```

```python
dist, dist2
```

```python
fig = plt.figure()
fig.patch.set_facecolor('0.2')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('0.2')

ax.scatter(A2[0], A2[1], A2[2], c='b', marker='o')
ax.scatter(B2[0], B2[1], B2[2], c='r', marker='o')
ax.plot(xs, ys, zs, c='black')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(0, 7)
ax.set_ylim(0, 7)
ax.set_zlim(0, 7)

plt.show()
```
