```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
data = 7 * np.random.randn(10000) + 180
```

```python
plt.figure(facecolor='grey',figsize=(5,3))
plt.axes().set_facecolor('0.3')
plt.hist(data,100, color='black')
plt.title('distribution of x')
plt.xlabel('x')
plt.ylabel('amount')
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 3),facecolor= '0.2')
for i in range(2):
    ax[i].set_facecolor('0.4')
    
ax[0].hist(data,np.arange(150,210,0.5), color = 'black',cumulative=True)
ax[0].set_title('cumulative observations < x')
ax[0].set_xlabel('x')
ax[0].set_ylabel('amount')
#ax[1].invert_xaxis()
ax[1].hist(data,np.arange(150,210,0.5), color = 'black', cumulative=-1)
ax[1].set_title('cumulative observations > x')
ax[1].set_xlabel('x')

plt.show()
```

```python
filter1 = data < 160
filter2 = (181 > data) & (data > 180)
filter3 = (data < 160) | (data > 200)

new_data2 = data[filter3]
new_data2.sort()
len(data[filter1]),len(data[filter3]), len(data[filter2])
```

```python
f = math.factorial
binomial_pmf = lambda n, p, x: (f(n) / (f(x) * f(n - x))) * (p ** x) * ((1 - p) ** (n - x))

n = 10
p = 0.5

x = np.arange(0, n+1)
y = [binomial_pmf(n, p, xi) for xi in x]

fig, ax = plt.subplots(facecolor='0.2')
ax.set_facecolor('black')

ax.bar(x, y, color='purple')

ax.set_xlabel("Number of successes")
ax.set_ylabel("Probability")
ax.set_title("Binomial Distribution (n = %d, p = %.2f)" % (n, p))

plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = list(map(lambda a: 1 / (1 + np.exp(-a)), x))

plt.figure(facecolor='.2',dpi = 100)
plt.axes().set_facecolor('.2')

plt.plot(x, y, color = 'black')
plt.xlabel("x")
plt.ylabel("Sigmoid(X)")

plt.show()
```
