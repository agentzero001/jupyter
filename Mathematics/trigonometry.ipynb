```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
custom_style = {
    'figure.facecolor': '.0',
    'axes.facecolor': '.1',
    'axes.labelcolor': 'white',
    'text.color': 'white',
    'xtick.color': 'green',
    'ytick.color': 'green',
    'axes.edgecolor': 'purple'
}
plt.rcParams.update(custom_style)


def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)
```

```python
x = np.linspace(0,10,10000)
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), c='grey')
ax.yaxis.set_major_locator(plt.MultipleLocator(1))
ax.set_ylim(-1.9,1.9);
```

```python
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=1, label='Sine',color='black')
ax.plot(x, np.cos(x), lw=1, label='Cosine', color='grey')

ax.grid(alpha=.3, axis='x')
ax.legend(frameon=False)
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
ax.yaxis.set_major_locator(plt.MultipleLocator(1))
ax.axhline(y=0,alpha=.2,color='grey')

ax.set_ylim(-1.9,1.9)
ax.set_xlim(0, 3 * np.pi);
```

```python
fig, ax = plt.subplots(figsize=(10,4))
x = np.linspace(0, 5 * np.pi, 10000)
ax.plot(x, np.sin(x), lw=1, label='Sine',color='black')
ax.plot(x, np.cos(x), lw=1, label='Cosine', color='grey')

ax.grid(alpha=.3, axis='x')
ax.legend(frameon=False)
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
ax.yaxis.set_major_locator(plt.MultipleLocator(1))
ax.axhline(y=0,alpha=.2,color='grey')

ax.set_ylim(-1.9,1.9)
ax.set_xlim(0, 5 * np.pi);
```

```python
fig, ax = plt.subplots(figsize=(10,3))
x = np.linspace(0, 5 * np.pi, 10000)
ax.plot(x, np.sin(x), lw=1, label='Sine',color='red', alpha=.6)

ax.grid(alpha=.2, axis='x')
ax.legend(frameon=False)
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
ax.yaxis.set_major_locator(plt.MultipleLocator(1))
ax.axhline(y=0,alpha=.2,color='grey')

ax.set_ylim(-1.9,1.9)
ax.set_xlim(0, 5 * np.pi);
plt.show()

fig, ax = plt.subplots(figsize=(10,3))
x = np.linspace(0, 5 * np.pi, 10000)
ax.plot(x, np.cos(np.pi/2-x), lw=1, label='Cos - pi/2',color='green',alpha=.7)

ax.grid(alpha=.2, axis='x')
ax.legend(frameon=False)
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
ax.yaxis.set_major_locator(plt.MultipleLocator(1))
ax.axhline(y=0,alpha=.2,color='grey')

ax.set_ylim(-1.9,1.9)
ax.set_xlim(0, 5 * np.pi);
plt.show()
```

In triangle ABC, angle C = 100, b = 4 cm, c = 5 cm. 
Find the magnitude of the angle B

Using the law of sines
## $\frac{\sin A}{a} = \frac{\sin B}{b} = \frac{\sin C}{c}$
solving for sin(B):
## $\sin(B) = \frac{b \cdot \sin(C)}{c}$
## $\sin(B) = \frac{4 \cdot \sin(100)}{5}$


```python
B = np.degrees(np.arcsin(4 * np.sin(d(100)) / 5))
print('B = {}'.format(round(B, 2)))
```
