```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.facecolor'] = '0.2'
```

```python
func = lambda x: x**2
d_func = lambda x: 2*x

f_sin = lambda x: np.sin(x)
f_cos = lambda x: np.cos(x)

x1 = np.linspace(np.pi, np.pi*3, 1000)
y_sin = f_sin(x1)
y_cos = f_cos(x1)


x2 = np.arange(-100,100,1)
f_x = func(x2)
df_x = d_func(x2)
```

```python
def calculate_positions(start, l_rate, func, d_func, n):
    positions = [(start, func(start))]
    curr_pos = (start, func(start))

    for _ in range(n - 1):
        new_x = curr_pos[0] - l_rate * d_func(curr_pos[0])
        new_y = func(new_x)
        curr_pos = (new_x, new_y)
        positions.append(curr_pos)
        
    return np.array(positions)
```

```python
A = calculate_positions(8,.1,func,d_func,15)
B = calculate_positions(6.5,.1,f_sin,f_cos,15)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(10,4))
ax1.set_facecolor('black')
ax2.set_facecolor('black')

ax1.plot(x2[100:110], f_x[100:110], c='green',zorder=1)
ax1.scatter(A[:,0], A[:,1],zorder=2,color='red', s=10)

ax2.plot(x1, y_sin, c='green',zorder=1)
ax2.scatter(B[:,0], B[:,1],zorder=2,color='red',s=10)
A
```

```python
c = lambda t: (np.sin(t), np.cos(t))
L = np.linspace(0,np.pi*2,100)

circle = list(map(c, L))
np_circle = np.array(circle)
plt.axes().set_facecolor('black')
plt.plot(np_circle[:,0], np_circle[:,1], color='green', linestyle='dotted')
plt.axis('equal')

plt.show()
```
