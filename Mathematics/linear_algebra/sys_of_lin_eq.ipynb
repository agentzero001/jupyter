```python
coefficients = np.array([[5,  1],
                         [2, -1]])

constants = np.array([3, 4])

np.linalg.solve(coefficients, constants)
```

```python
X, Y = np.meshgrid(np.arange(-5, 5, 1),np.arange(-5, 5, 1))
Z1 = list(map(lambda x, y: -x/2 + -y/2 + 4.5, X, Y))
Z2 = list(map(lambda x, y: 2/3 * x + 4/3 * y - 1/3, X, Y))
Z3 = list(map(lambda x, y: .6 * x + 1.2 * y, X, Y))

Z1, Z2, Z3 = np.array(Z1), np.array(Z2), np.array(Z3)
```

```python
coeffs = np.array([[1,1,2],
                   [2,4,-3],
                   [3,6,-5]])

const = np.array([9,1,0])

np.linalg.solve(coeffs, const)
```

```python
fig = plt.figure(facecolor='.2')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('.2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.set_xlim(-10, 10)
ax.set_ylim(-10, 10)
ax.set_zlim(-10, 10)

ax.plot([-10, 10], [0, 0], [0, 0], color='black', linestyle='--')
ax.plot([0, 0], [-10, 10], [0, 0], color='black', linestyle='--')
ax.plot([0, 0], [0, 0], [-10, 10], color='black', linestyle='--')


ax.plot_surface(X, Y, Z1, alpha=0.5, color='red', zorder=0)
ax.plot_surface(X, Y, Z2, alpha=0.5, color='green',zorder=1)
ax.plot_surface(X, Y, Z3, alpha=0.5, color='yellow',zorder=2)
plt.show()
```
