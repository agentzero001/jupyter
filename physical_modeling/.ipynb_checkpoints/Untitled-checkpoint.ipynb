```python
N = np.arange(0, 30, .1)
g = lambda n, μ: np.e**(-μ) * μ**n / factorial(n)    
Xs = [g(N, mu) for mu in range(1, 10)]

for xs in Xs:
    plt.plot(N, xs)
```

```python
initial_speed = 0.0
impact_time = 0.0
g = 9.80665
initial_height = 2.0
speed_increment = 5.0 
cutoff_time = 10.0

while impact_time < cutoff_time:
    impact_time = (np.sqrt(initial_speed**2 + 2 * g * initial_height) + initial_speed) / g
    print("speed = {} m/s; time = {:.1f} s".format(initial_speed, impact_time))
    initial_speed += speed_increment
print("calculation complete.")
```

```python
x = np.linspace(0, 1, 1001)
y = 3*np.sin(x)**3 - np.sin(x)
```

```python
x = np.arange(0, 10, .1)

plt.plot(x, x, label="Population 1")
plt.plot(x, x**2, label="Population 2")
plt.legend()
ax = plt.gca()
lines = ax.get_lines()
lines[0].set_label("Infected Population")
lines[1].set_label("Cured population")
ax.legend()
```

```python
#give plt.plot an ordinary vector of x values, but a two dimensional array of y values.
#each column of y will be used to draw a separate curve using a commmon set of x values.

num_curves = 3
x = np.linspace(0, 1, 51)
y = np.zeros( (x.size, num_curves) )
for n in range(num_curves):
    y[:, n] = np.sin((n + 1) * x * 2 * np.pi)
plt.plot(x, y)
```

```python
from numpy.random import random
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 101)
plt.figure()
plt.subplot(2, 2, 1); plt.hist(random(20))
plt.subplot(2, 2, 2); plt.plot(t, t**2, t, t**3 - t)
plt.subplot(2, 2, 3); plt.plot(random(20), random(20), 'r*')
plt.subplot(2, 2, 4); plt.plot(t*np.cos(10*t), t*np.sin(10*t))
plt.suptitle("Data and Functions")

fig = plt.gcf()
fig.canvas.get_supported_filetypes()
plt.savefig("greatest_figure_ever.pdf")
```
