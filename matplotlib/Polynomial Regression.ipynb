```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1= np.round( np.random.normal(1.75, 0.20, 700), 2)
y1= x1**5 * np.round( np.random.normal(60.32, 10, 700), 2)
#np_city=np.column_stack((height, weight))

plt.figure(facecolor='grey',dpi = 100)
plt.axes().set_facecolor('grey')

plt.scatter(x1,y1, color='black', s=5)
#plt.plot(x2,y2, color ='red')

plt.show()
```

```python
def lift(arr):
    arr = np.array(arr)
    return arr.reshape((arr.shape[0],1))
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit([[i**3] for i in x1], lift(y1))
```

```python
x2 = np.linspace(1,2.5,10)
y2 = model.predict([[i**3] for i in x2])

plt.figure(facecolor='grey',dpi = 100)
plt.axes().set_facecolor('grey')

plt.scatter(x1,y1, color='black', s=3)
plt.plot(x2,y2, color ='red')
```
