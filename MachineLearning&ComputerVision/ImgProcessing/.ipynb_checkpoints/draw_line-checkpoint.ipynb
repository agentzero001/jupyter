```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.facecolor'] = '0.3'
```

```python
def line(A, x1, y1, x2, y2):
    d = (y2 - y1) / (x2 - x1)
    for x in range(x1, x2 + 1):
        A[round(d * (x - x1) + y1), x] = 1

B = np.zeros((40, 40))
C = B.copy()
for i in range(1,20):
    B[round(0.5*i), i] = 1

C[[round(0.5*i) for i in range(1, 20)], [i for i in range(1, 20)]] = 1

A_ = np.array([[1 if j < 20 and i == 10 else 0 for j in range(20)] for i in range(20)])
```

```python
plt.matshow(C, cmap='gray', origin='lower')
```

```python
def line_(arr, x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    err = dx / 2
    while True:
        arr[y1, x1] = 1
        if x1 == x2 and y1 == y2:
            break
        x1 += 1
        err -= dy
        if err < 0:
            y1 += 1
            err += dx
```

```python
D = np.zeros((50,50))
line_(D, 1, 2, 20, 20)
plt.matshow(D, cmap='gray', origin='lower')
```

```python
#Bresenham algorithm
def line__(arr, x1, y1, x2, y2):
    dx = abs(x2 - x1)
    sx = 1 if x1 < x2 else -1
    dy = -abs(y2 - y1)
    sy = 1 if y1 < y2 else -1
    err = dx + dy
    while True:
        arr[y1, x1] = 1
        if x1 == x2 and y1 == y2:
            break
        e2 = 2 * err
        if (e2 > dy):
            err += dy
            x1 += sx
        if (e2 < dx):
            err += dx
            y1 += sy
```

```python
E = np.zeros((50,50))
line__(E, 1, 1, 20, 40)
line__(E, 1, 40, 20, 1)
plt.matshow(E, cmap='gray', origin='lower')
```

```python
def rectangle(arr, a, b, c):
    d = (c[0] + a[0] - b[0], c[1] + a[1] - b[1])
    line__(arr, *a, *b)
    line__(arr, *b, *c)
    line__(arr, *c, *d)
    line__(arr, *d, *a)
```

```python
L3 = np.zeros((25,25))
rectangle(L3, (6,2), (21,12), (17,18))
```

```python
plt.matshow(L3, cmap='gray', origin='lower')
```
