```python
x = np.linspace(-1.5,1.5,200)
x, y = np.meshgrid(x, x)
z = np.exp(-x*x-y*y)

fig = plt.figure(facecolor='grey')
ax = fig.gca(projection = '3d')
ax.set_facecolor('grey')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(x,y,z, cmap='Spectral')
```

```python
picture = np.zeros((200,200))
for x in range(200):
    for y in range(200):
        if x > 50 and y > 50 and y < 200 - x:
            picture[x,y] = 1
plt.figure(facecolor='0.2')
plt.imshow(picture,cmap='gray')
```

```python
filterX = np.zeros((3,3))
filterX[1,0] = -0.5
filterX[1,2] = 0.5

filterY = np.zeros((3,3))
filterY[0,1] = -0.5
filterY[2,1] = 0.5

filterX, filterY
```

```python
from scipy.signal import convolve2d
dx = convolve2d(picture, filterX)
dy = convolve2d(picture, filterY)
```

```python
fig,ax = plt.subplots(1,2, facecolor='0.2',figsize=(8,6))
ax[0].imshow(dx,cmap='gray')
ax[1].imshow(dy,cmap='gray')
```

```python
dx[45:60,45:55] #notice convolve2d will move the convolutionKernel in the opposite direction
```

```python
plt.figure(facecolor='0.2')
plt.imshow(np.sqrt(dx*dx+dy*dy), cmap='gray')
```

```python
pic = Image.open('mypic2.jpg').convert('L')
```

```python
plt.figure(facecolor='0.2')
plt.imshow(pic, cmap='gray') #oh it's you
```

```python
#gradient edge detection
filterX = np.zeros((3,3))
filterY = np.zeros((3,3))
filterX[1,0] = -0.5
filterX[1,2] = 0.5
filterY[0,1] = -0.5
filterY[2,1] = 0.5
dx1 = convolve2d(pic, filterX)
dy1 = convolve2d(pic, filterY)
edges = np.sqrt(dx1*dx1+dy1*dy1)
```

```python
#Roberts cross
mx = np.array([[1, 0], [0, -1]])
my = np.array([[0, 1], [-1, 0]])
Gx = convolve2d(pic, mx)
Gy = convolve2d(pic, my)
edges2 = np.abs(Gx) + np.abs(Gy)
```

```python
fig,ax = plt.subplots(1,2, facecolor='0.2',figsize=(10,8))
ax[0].imshow(edges, cmap='gray')
ax[0].set_title('Gradient edge detection')
ax[1].imshow(edges2,cmap='gray')
ax[1].set_title('Roberts cross')
plt.show()
```
