```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import warnings
warnings.filterwarnings('ignore')
```

```python
%matplotlib notebook
#%matplotlib inline
```

```python
x = np.linspace(1,60,200)

fig = plt.figure(facecolor='grey')
ax = fig.gca(projection = '3d')
ax.set_facecolor('grey')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')


ax.scatter(np.sin(x),np.cos(x),x/2,c='black')
ax.scatter([4],[4],[4], c='red')
plt.show()
```

```python
x = np.linspace(-1.5,1.5,200)
x, y = np.meshgrid(x, x)
z = np.exp(-x*x-y*y)

fig = plt.figure(facecolor='grey')
ax = fig.gca(projection = '3d')
ax.set_facecolor('grey')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(x,y,z, cmap='Spectral')
```
