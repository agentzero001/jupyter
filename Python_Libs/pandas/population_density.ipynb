```python
pop = pd.read_csv('data/state-population.csv')
areas = pd.read_csv('data/state-areas.csv')
abbrevs = pd.read_csv('data/state-abbrevs.csv')
```

```python
display(pop.head(), areas.head(), abbrevs.head())
```

```python
areas['state'].value_counts()
```

```python
merged = pd.merge(pop, abbrevs, how='outer',
                  left_on='state/region', right_on='abbreviation')
merged.head()
```

```python
merged.drop('abbreviation', axis=1, inplace=True)
merged.head()
```

```python
merged.isnull().any()
```

```python
merged[merged['population'].isnull()]
```

```python
merged[merged['population'].isnull()]['state/region'].unique()
```

```python
merged[merged['state'].isnull()]
```

```python
merged[merged['state'].isnull()]['state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
```

```python
merged.isnull().any()
```

```python
final = pd.merge(merged, areas, on='state', how='left')
final.head()
```

```python
final.isnull().any()
```

```python
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
final.dropna(inplace=True)
```

```python
final.to_csv('data.csv', index=False)
final_ = pd.read_csv('data.csv')
final_.head()
```

```python
data2010 = final_.query("year == 2010 & ages == 'total'")
data2010.head()
```

```python
(final_[(final_['year'] == 2010) & (final_['ages'] == 'total')]
 == data2010).all() #ok (just to make sure)
```

```python
data2010.set_index('state', inplace=True)
density = data2010['population'] / data2010['area (sq. mi)']
density.sort_values(ascending=False, inplace=True)
density.head()
```
