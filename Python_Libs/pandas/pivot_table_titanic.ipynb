```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
pd.pivot_table(titanic,
    values='survived',
    index='sex',
    columns='class',
    aggfunc = 'mean',
    fill_value=None,
    margins = False,
    dropna = True,
    margins_name = 'All',
    observed = False,
    sort = True
)
```

```python
df = pd.DataFrame(titanic)
display(df.head(), df.info())
```

```python
titanic.groupby('sex')[['survived']].mean()
```

```python
t1 = titanic.groupby(['sex', 'class'])['survived'].aggregate('mean').unstack()
t2 = titanic.pivot_table('survived', index='sex', columns='class')
display(t1, t2)
```

```python
age = pd.cut(titanic['age'], [0, 18, 80])
titanic.pivot_table(['survived'], ['sex', age], 'class')
```

```python
titanic.pivot_table(['survived', 'fare'], ['sex', age], 'class')
```

```python
fare = pd.qcut(titanic['fare'], 2)
titanic.pivot_table('survived', ['sex', age], [fare, 'class'])
```

```python
titanic.pivot_table(index='sex', columns='class',
                    aggfunc={'survived':sum, 'fare':'mean'})
```

```python
titanic.pivot_table(index='sex', columns='class', 
                    values=['survived', 'fare'],
                    aggfunc={'survived': sum, 'fare': 'mean'}, 
                    margins=True)
```

```python
titanic.pivot_table('survived', index='sex', columns='class', margins=True)
```
