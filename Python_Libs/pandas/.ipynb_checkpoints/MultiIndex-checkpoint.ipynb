```python
import pandas as pd
import numpy as np
```

```python
f = lambda s: np.random.randint(low=10**7, high=10**8, size=s)
```

```python
f(5), f((2,5))
```

```python
pd.MultiIndex(levels=[['a','b'],[0,1,2]], codes=[[0,1,0,1],[2,1,0,2]])
```

```python
pd.MultiIndex.from_tuples([('a', 5, 2),('a',3),('b', 4)])
```

```python
index = pd.MultiIndex.from_product([{'A','B','C','D'},{1,2,3}], names=['first','second'])
index
```

```python
S = pd.Series(f(12), index=index)
S
```

```python
S.index.names = [None, None]
S
```

```python
df = S.unstack(level=-1)
df
```

```python
S = df.stack()
```

```python
S.index.names = ['first', 'second']
S
```

```python
df = pd.DataFrame(S,columns=['population'])
```
