```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

```python
custom_style = {
    'figure.facecolor': '.0',
    'axes.facecolor': '.2',
    'axes.labelcolor': 'white',
    'text.color': 'white',
    'xtick.color': 'green',
    'ytick.color': 'green',
    'axes.edgecolor': 'purple'
}
sns.set_style("dark", rc=custom_style)
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    plt.hist(data[col], alpha=0.5)
```

```python
for col in 'xy':
    sns.kdeplot(data[col], fill=True)
```

```python
sns.histplot(data=data['x'], kde=True, label='x', color='orange')
sns.histplot(data=data['y'], kde=True, label='y', color='yellow')
```

```python
sns.jointplot(data=data, x='x', y='y', kind='kde', cmap="magma")
```
