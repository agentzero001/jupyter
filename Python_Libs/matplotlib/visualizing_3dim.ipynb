```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.rcParams['figure.facecolor'] = '.0'
plt.rcParams['axes.facecolor'] = '.1'
plt.rcParams['axes.labelcolor'] = 'white'
plt.rcParams['text.color'] = 'white'
plt.rcParams['xtick.color'] = 'green'
plt.rcParams['ytick.color'] = 'green'
plt.rcParams['axes.edgecolor'] = 'purple'
```

```python
f = lambda x, y: np.sin(x) ** 20 + np.cos(y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.contour(X, Y, Z, colors='grey')

plt.subplot(1,2,2)
plt.contour(X, Y, Z, 20, cmap='inferno', alpha=.7);
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```

```python
f = lambda x, y: np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-12, 12, 60)
y = np.linspace(-12, 12, 60)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
fig = plt.figure(figsize=(8,8))
fig.patch.set_facecolor('.2')

ax1, ax2, ax3, ax4 = [fig.add_subplot(2,2,i, projection='3d') for i in range(1,5)]
titles = ['contour3D', 'countour3D', 'wireframe', 'surface']

for ax, title in zip((ax1,ax2,ax3,ax4),titles):
    ax.set_facecolor('.2')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')
    ax.set_title(title)

ax1.contour3D(X, Y, Z, 50, cmap='binary')

ax2.contour3D(X, Y, Z, 50, cmap='binary')
ax2.view_init(60, 35)

ax3.plot_wireframe(X, Y, Z, color='black')

ax4.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none');
```

```python
r = np.linspace(0, 6, 20)
theta = np.linspace(-0.9 * np.pi, 0.8 * np.pi, 40)
r, theta = np.meshgrid(r, theta)

X = r * np.sin(theta)
Y = r * np.cos(theta)
Z = f(X, Y)

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z,
                rstride=1,
                cstride=1,
                cmap='viridis',
                edgecolor='none');
```

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)
```

```python
fig = plt.figure(figsize=(10,5))

ax = [fig.add_subplot(1, 2, 1, projection='3d'),
      fig.add_subplot(1, 2, 2, projection='3d')]

ax[0].scatter(x, y, z, c=z, cmap='viridis', linewidth=0.5)
ax[1].plot_trisurf(x, y, z, cmap='viridis', edgecolor='none')
```

```python
#mbius strip

theta = np.linspace(0, 2 * np.pi, 30)
w = np.linspace(-0.25, 0.25, 8)
w, theta = np.meshgrid(w, theta)

phi = 0.5 * theta

r = 1 + w * np.cos(phi)

x = np.ravel(r * np.cos(theta))
y = np.ravel(r * np.sin(theta))
z = np.ravel(w * np.sin(phi))

from matplotlib.tri import Triangulation
tri = Triangulation(np.ravel(w), np.ravel(theta))

ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles,
                cmap='viridis', linewidths=0.2);

#ax.view_init(70,80)
ax.set_xlim(-1, 1); ax.set_ylim(-1, 1); ax.set_zlim(-1, 1);
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.tri import Triangulation

# Parameters
u = np.linspace(0, 2 * np.pi, 30)
v = np.linspace(0, np.pi, 30)
u, v = np.meshgrid(u, v)

# Parametric equations for a sphere
x = np.ravel(np.sin(v) * np.cos(u))
y = np.ravel(np.sin(v) * np.sin(u))
z = np.ravel(np.cos(v)) 

# Triangulation
tri = Triangulation(np.ravel(u), np.ravel(v))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap='magma', linewidth=0.2)

ax.set_xlim(-2, 3)
ax.set_ylim(-2, 3)
ax.set_zlim(-2, 3)

plt.show()
```
