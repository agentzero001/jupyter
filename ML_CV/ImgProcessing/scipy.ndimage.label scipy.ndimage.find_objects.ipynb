```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from PIL import Image
plt.rcParams['figure.facecolor'] = '0.2'
```

```python
def more_plot(pics, shape, c='jet', s=(8,8), sh=False):
    fig, axs = plt.subplots(shape[0], shape[1], facecolor='0.2',figsize=s, sharey=sh)
    for idx, val in enumerate(pics):
        i = idx // shape[1]
        j = idx % shape[1]
        axs[i,j].imshow(pics[idx], cmap=c)
    plt.show()
```

```python
def nplot(pics ,c='gray', s=(10,8),sh=False):
    fig, axs = plt.subplots(1, len(pics), facecolor='0.2', figsize=s, sharey=sh)
    for i in range(len(pics)):
        axs[i].imshow(pics[i], cmap=c)
    plt.show()
```

```python
def discreteImage(M):
    max_ = np.max(M)
    min_ = np.min(M)
    cmap = plt.get_cmap('jet', max_ - min_ + 1)
    plt.matshow(M, cmap=cmap, vmin = min_ - 0.5, vmax = max_ + 0.5)
    plt.colorbar(ticks=np.arange(min_, max_ + 1))
```

```python
a = np.random.randint(1, 4, size=(4,4))
my_cmap = plt.get_cmap('jet', 3)
plt.matshow(a, cmap=my_cmap, vmin=0.5, vmax=3.5)
plt.colorbar(ticks=range(1,4))
```

```python
from scipy.ndimage import label
from scipy.ndimage import find_objects
```

```python
image = np.zeros((9, 9))
image[2:4, 3:6] = 1
image[5:8, 1:4] = 1

labeled_img = label(image)[0]
slices = find_objects(labeled_img)
obj = [labeled_img[slices[0]], labeled_img[slices[1]]]
```

```python
nplot([image, labeled_img, obj[0], obj[1]], c='jet')
```

```python
img1 = np.array(Image.open('mypic3.png').convert('L'))
img2 = img1[450:800,200:1340]
img3 = ((img2 < 100)*1)
```

```python
np.unique(img3, return_counts=True)
discreteImage(img3)
```
