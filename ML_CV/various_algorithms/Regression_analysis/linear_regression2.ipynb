```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.facecolor'] = '0.3'
```

```python
np.random.seed(42)

x1 = np.random.randint(10,100, size=(20))
y1 = x1 + np.random.randint(1,50, size=(20))

x1, y1
```

```python
plt.figure(facecolor='.2',dpi = 100)
plt.axes().set_facecolor('black')
plt.axis([0,110,0,120])
plt.scatter(x1,y1, color='green', alpha=.5)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

def lift(arr):
    arr = np.array(arr)
    return arr.reshape((arr.shape[0], 1))

model = LinearRegression()
model.fit(lift(x1), lift(y1))
```

```python
# model.get_params()
# model.__dict__
model.predict([[40]])
```

```python
model.coef_[0][0], model.intercept_[0]
```

```python
plt.figure(facecolor='.2',dpi = 100)
plt.axes().set_facecolor('black')
plt.axis([0,110 ,0, 120])
plt.scatter(x1,y1, color='green',zorder=3)
plt.plot([1, 120], model.predict([[1],[120]]), color ='red',zorder=2)

plt.grid(zorder=1, alpha=.2)
plt.show()
```


$ r(a, b) = X_{\text{sq}} \cdot a^2 + 2 \cdot \sum_{i=1}^{n} x_{1i} \cdot a \cdot b + n \cdot b^2 - 2 \cdot XY \cdot a - 2 \cdot \sum_{i=1}^{n} y_{1i} \cdot b + Y_{\text{sq}} $



```python
m_, b_
```

```python
Xsq = np.sum(x1*x1)
Ysq = np.sum(y1*y1)
XY = np.sum(x1*y1)
n = len(x1)

a, b = np.meshgrid(np.linspace(1.1, 1.3, 1000), np.linspace(12, 20, 1000))
r = (Xsq * a * a) + (2 * np.sum(x1) * a * b) + (n * b * b) - (2 * XY * a) - (2 * np.sum(y1) * b) + Ysq

fig = plt.figure(facecolor='.1')
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('.1')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.plot_surface(a, b, r, color = 'black',alpha=.8)
ax.scatter(m_, b_, r.min(), color='red', s=70)
plt.show()
```

```python
#%matplotlib notebook
```

```python
#This function returns the index of the minimum value when the array is flattened.
r.argmin()
```

```python
#This function converts a flattened index back into a tuple of indices in the original array shape.
min_index = np.unravel_index(r.argmin(), r.shape)
min_index
```

```python
r[min_index], r.min()
```

```python
a[min_index], m_, b[min_index], b_
```

```python
#in the cells below you can see that the optimal values of a and b are approx 0.993 and 10.085 respectively
#this is just another way of visualizing 3dim.
a, b = np.meshgrid(np.linspace(1.1, 1.3, 1000), np.linspace(15, 18, 1000))
r = (Xsq * a * a) + (2 * np.sum(x1) * a * b) + (n * b * b) - (2 * XY * a) - (2 * np.sum(y1) * b) + Ysq

plt.imshow(r, cmap='viridis', extent=(0, 2, 15, 17), aspect='auto')
plt.scatter(m_, b_, color='black')
plt.colorbar()
plt.xlabel('a')
plt.ylabel('b')
plt.title('Parameter Space Visualization')
plt.show()
```
